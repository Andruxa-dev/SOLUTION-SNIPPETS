```python
def tksheets_function():
    
    class demo(tk.Tk):
        
        def __init__(self):
            
            tk.Tk.__init__(self)
            self.grid_columnconfigure(0, weight = 1)
            self.grid_rowconfigure(0, weight = 1)
            self.frame = tk.Frame(self)
            self.frame.grid_columnconfigure(0, weight = 1)
            self.frame.grid_rowconfigure(0, weight = 1)
            self.sheet = Sheet(self.frame, align = "center", data =[[f'' for c in range (13)] for r in range (70)], theme = "dark green", height = 300, width = 900)
            self.sheet.sheet_data_dimensions(total_rows = 69, total_columns = 13)
            self.sheet.change_theme("dark")
            self.sheet.enable_bindings()
            self.frame.grid(row = 0, column = 0, sticky = "nswe")
            self.sheet.grid(row = 0, column = 0, sticky = "nswe")
            
            
            #Fixed/variable locations on the sheet to assign locations of names, variables / user input###############################################################
            column1_fixed_map ={'user_data1':[10,1],'user_data2':[11,1],'user_data3':[12,1],'user_data4':[13,1],'user_data5':[14,1]}
            
            for key, values in column1_fixed_map.items():
                
                self.sheet.set_cell_data(values[0], values[1],key)
                
            #################### 
            a = 0
            column1_variable_map ={'user_data1':[10,2],'user_data2':[11,2],'user_data3':[12,2],'user_data4':[13,2],'user_data5':[14,2]}
            
            for key1, values1 in column1_variable_map.items():
                
                self.sheet.set_cell_data(column1_variable_map[key1][0], column1_variable_map[key1][1], user_input_list1[a])
                                         
                a+=1
                
            #Assignment to locations of 'excell data'##################################################################################################################
            
            self.sheet.set_cell_data(9, 3, "excell_data/date")# getting one fixed location first "Column name"
            self.sheet.highlight_cells( row = 9, column = 3, bg = "cyan", fg = "black")
            
            ####################
                        
            column2_variable_map ={'excel_data1':[10,3],'excel_data2':[11,3],'excel_data3':[12,3],'excel_data4':[13,3],'excel_data5':[14,3]}
            
            for key2, values2 in column2_variable_map.items():
                
                for key3 in excel_data1_global:
                    
                    if key2 == key3:
                        
                        key2 = excel_data1_global[key3]
                        self.sheet.set_cell_data(values2[0],values2[1], key2)
                    
                    else:
                        #if there is a key mismatch between two dictionaries
                        pass
                        
            #Operation between columns of tksheets just to show functionality#######################################################################################
            #column will be fixed at 3 another clumn at 2, result will be an addition  between two
            
            self.sheet.set_cell_data(9, 4, "addition result")# getting one fixed location first "Column name"
            self.sheet.highlight_cells( row = 9, column = 4, bg = "cyan", fg = "black")
            
            row_loc_list  = [10,11,12,13,14]
            for i in row_loc_list:
                
                aa = float(self.sheet.get_cell_data(i, 3)) 
                bb = float(self.sheet.get_cell_data(i, 2))
                xx = aa + bb
                self.sheet.set_cell_data(i, 4, xx)
               
    p = demo()
    p.mainloop()
```
